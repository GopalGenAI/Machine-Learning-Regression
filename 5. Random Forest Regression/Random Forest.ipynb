```python
import pandas as pd #for handling the tables
```

```python
dataset = pd.read_csv("50_Startups.csv") #Dataset collection
```

```python
#Dataset preprocessing
dataset = pd.get_dummies(dataset, drop_first = True, dtype = int)
#convert the classification data into numerical data
```

```python
#get the columns
dataset.columns
```

```python
independent = dataset[['R&D Spend', 'Administration', 'Marketing Spend', 'State_Florida', 'State_New York' ]]
dependent = dataset [['Profit']]
#split the input and output
```

```python
#Split train and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
#create the model
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=100, random_state = 0)
regressor = regressor.fit(x_train, y_train)
```

```python
y_pred = regressor.predict(x_test)
y_pred # from the trained model predict the test input 
```

```python
from sklearn.metrics import r2_score #evaluate the model by r2_score
r_score = r2_score(y_test, y_pred)
r_score
```

```python
import pickle #save the best model in file by naming it
filename = "finalmodel_Random_forest.sav"
pickle.dump(regressor, open(filename, 'wb'))
```

```python
loadedmodel = pickle.load(open("finalmodel_Random_forest.sav", 'rb'))
result = loadedmodel.predict([[13, 14, 5, 1, 0]])
result #profit of 5 input values
```
