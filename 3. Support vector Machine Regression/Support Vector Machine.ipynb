```python
import pandas as pd #import library
```

```python
dataset = pd.read_csv("50_Startups.csv")
# data collection
```

```python
dataset = pd.get_dummies(dataset, drop_first = True, dtype = int)
#data preprocessing 
```

```python
#seperate input and output
dataset.columns
```

```python
independent = dataset[['R&D Spend', 'Administration', 'Marketing Spend', 
       'State_Florida', 'State_New York']]
dependent = dataset [['Profit']]
```

```python
#seperate train and test set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(independent, dependent, test_size=0.30, random_state =0)
```

```python
from sklearn.svm import SVR #call support vector machine model
regressor = SVR(kernel = "linear", C=0.1)
regressor = regressor.fit (x_train, y_train)
regressor
```

```python
regressor.coef_ #slope/weight
```

```python
regressor.n_support_ #it represent number of data points
```

```python
regressor.support_
```

```python
regressor.support_vectors_ #set the inputs within range (-1 to 1)
```

```python
y_pred = regressor.predict(x_test) #from the trained model regressor predict the test input 
y_pred
```

```python
from sklearn.metrics import r2_score #evaluate the model
rscore = r2_score(y_test, y_pred)
rscore
```

```python
import pickle
filename = "finalmodel_SVM.sav"
pickle.dump(regressor, open("filename", 'wb'))
```
