```python
import xgboost
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
dataset = pd.read_csv('insurance_pre.csv')
#dataset collection
```

```python
#dataset preprocessing
dataset = pd.get_dummies(dataset, drop_first = True, dtype = int)
#convert the classification data into numerical data
```

```python
dataset.columns
```

```python
#seperate the input and output
independent = dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes' ]]
dependent = dataset [['charges']]
```

```python
#seperate train and test da_taset
x_train, x_test, y_train, y_test = train_test_split(independent, dependent, test_size = 30,random_state = 0)
```

```python
from xgboost import XGBRegressor
regressor = XGBRegressor(n_estimators=1000, max_depth=5, eta=0.01, subsample=0.6, colsample_bytree=0.9) 
regressor.fit (x_train, y_train)
regressor
```

```python
y_pred = regressor.predict(x_test)
y_pred
```

```python
r_score = r2_score(y_test, y_pred)
r_score
```

```python
import pickle
filename = 'finalized_model_xgbt.sav'
pickle.dump(regressor,open('finalized_model_xgbt.sav', 'wb'))
```
