```python
import pandas as pd
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import GridSearchCV
```

```python
dataset = pd.read_csv('insurance_pre.csv')
#dataset collection
```

```python
#dataset preprocessing
dataset = pd.get_dummies(dataset, drop_first = True, dtype = int)
#convert the classification data into numerical data
```

```python
dataset.columns
```

```python
independent = dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes' ]]
dependent = dataset [['charges']]
```

```python
parameters = {
    'loss': ['squared_error', 'absolute_error', 'huber', 'quantile'],
    'learning_rate': [0.1],
    'n_estimators': [100],
    'subsample': [1.0],
    'criterion': ['friedman_mse', 'squared_error'],
    'min_samples_split': [2],
    'min_samples_leaf': [1],
    'min_weight_fraction_leaf': [0.0],
    'max_depth': [3],
    'min_impurity_decrease': [0.0],
    'init': [None],
    'random_state': [None],
    'max_features': [None],
    'alpha': [0.9],
    'verbose': [0],
    'max_leaf_nodes': [None],
    'warm_start': [False],
    'validation_fraction': [0.1],
    'n_iter_no_change': [None],
    'tol': [0.0001],
    'ccp_alpha': [0.0]
}
grid = GridSearchCV(GradientBoostingRegressor(), parameters, refit = True, verbose = 3, n_jobs = -1)
grid.fit(independent, dependent)
```

```python
result = grid.cv_results_
print ("The best R_score value for parameter {}.".format(grid.best_params_))
```

```python
table = pd.DataFrame.from_dict(result)
table
```
