```python
import pandas as pd
from sklearn.ensemble import AdaBoostRegressor
from sklearn.model_selection import GridSearchCV
```

```python
dataset = pd.read_csv('insurance_pre.csv')
#dataset collection
```

```python
#dataset preprocessing
dataset = pd.get_dummies(dataset, drop_first = True, dtype = int)
#convert the classification data into numerical data
```

```python
dataset.columns
```

```python
independent = dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes' ]]
dependent = dataset [['charges']]
```

```python
param_grid = {
    'estimator': [None],
    'n_estimators': [50, 100, 150],
    'learning_rate': [1.0],
    'loss': ['linear', 'square', 'exponential'],
    'random_state': [None]
}
grid = GridSearchCV(AdaBoostRegressor(), param_grid, refit = True, verbose = 3, n_jobs =-1)
grid.fit (independent, dependent)
```

```python
result = grid.cv_results_
print("The best R_Score value for parameter {}.". format(grid.best_params_))
```

```python
table = pd.DataFrame.from_dict(result)
table
```

```python
import pickle
filename = "finalized_Model_Adb.sav"
pickle.dump(grid, open('finalized_Model_Adb.sav', 'wb'))
```
